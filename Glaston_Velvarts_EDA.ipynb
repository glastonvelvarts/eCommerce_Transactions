Merging 2 dataframes based on Customer and Product ID

```python
# Merge df1 and df2 on CustomerID
merged_df = pd.merge(df1, df2, on='CustomerID')
```

```python
merged_df
```

```python
# Group by Region and sum the TotalValue
region_sales = merged_df.groupby('Region')['TotalValue'].sum().reset_index()
# Find the region with the highest total sales
highest_sales_region = region_sales.loc[region_sales['TotalValue'].idxmax()]

print("Region with the highest total sales:")
print(highest_sales_region)
```

```python
# Group by ProductID and sum the Quantity
product_sales = merged_df.groupby('ProductID')['Quantity'].sum().reset_index()
# Sort the products by Quantity in descending order
most_purchased_products = product_sales.sort_values(by='Quantity', ascending=False)

print("Most purchased products:")
print(most_purchased_products.head())

# Merge with df3 to get the Category information
product_sales_with_category = pd.merge(most_purchased_products, df3[['ProductID', 'Category']], on='ProductID')

# Group by Category and sum the Quantity
category_sales = product_sales_with_category.groupby('Category')['Quantity'].sum().reset_index()
# Sort the categories by Quantity in descending order
most_purchased_categories = category_sales.sort_values(by='Quantity', ascending=False)

print("\nMost purchased categories:")
print(most_purchased_categories.head())
```

```python
# Ensure TransactionDate is in datetime format
merged_df['TransactionDate'] = pd.to_datetime(merged_df['TransactionDate'], errors='coerce')

# Drop rows with invalid TransactionDate or TotalValue
merged_df = merged_df.dropna(subset=['TransactionDate', 'TotalValue'])

# Extract year and month from TransactionDate
merged_df['Year'] = merged_df['TransactionDate'].dt.year
merged_df['Month'] = merged_df['TransactionDate'].dt.month

# Group by Year and Month, summing TotalValue
monthly_revenue = merged_df.groupby(['Year', 'Month'], as_index=False)['TotalValue'].sum()

# Group by Year, summing TotalValue
yearly_revenue = merged_df.groupby('Year', as_index=False)['TotalValue'].sum()

# Display results
print("Monthly Revenue Trends:")
print(monthly_revenue)

print("\nYearly Revenue Trends:")
print(yearly_revenue)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensuring months are from (January to December)
monthly_revenue['Month'] = pd.Categorical(
    monthly_revenue['Month'],
    categories=range(1, 13),  #this here is the range of the months
    ordered=True
)

plt.figure(figsize=(16, 8))

# Plot monthly revenue trends
plt.subplot(1, 2, 1)
sns.lineplot(
    data=monthly_revenue,
    x='Month',
    y='TotalValue',
    hue='Year',
    marker='o',
    palette='tab10'
)
plt.title('Monthly Revenue Trends', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Plot yearly revenue trends
plt.subplot(1, 2, 2)
sns.barplot(
    data=yearly_revenue,
    x='Year',
    y='TotalValue',
    palette='pastel'
)
plt.title('Yearly Revenue Trends', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add annotations for the bar chart
for index, row in yearly_revenue.iterrows():
    plt.text(
        x=index,
        y=row['TotalValue'] + (0.01 * yearly_revenue['TotalValue'].max()),  # Slightly above the bar
        s=f"{row['TotalValue']:.2f}",
        ha='center',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

```python
# Group by CustomerID and sum the TotalValue
customer_revenue = merged_df.groupby('CustomerID')['TotalValue'].sum().reset_index()

# Sort the customers by TotalValue in descending order
high_value_customers = customer_revenue.sort_values(by='TotalValue', ascending=False)

print("High-Value Customers:")
print(high_value_customers.head())
```

```python
# Ensure SignupDate is in datetime format
df1['SignupDate'] = pd.to_datetime(df1['SignupDate'], errors='coerce')

# Extract month from SignupDate
df1['SignupMonth'] = df1['SignupDate'].dt.month

# Count the number of signups per month
signup_counts = df1['SignupMonth'].value_counts().sort_index()

# Plot the signup patterns
plt.figure(figsize=(10, 6))
sns.barplot(x=signup_counts.index, y=signup_counts.values, palette='viridis')
plt.title('Peak Months for Customer Signups', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Signups', fontsize=12)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from fpdf import FPDF

class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'Business Insights', 0, 1, 'C')

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial', 'I', 8)
        self.cell(0, 10, f'Page {self.page_no()}', 0, 0, 'C')

    def chapter_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, title, 0, 1, 'L')
        self.ln(10)

    def chapter_body(self, body):
        self.set_font('Arial', '', 12)
        self.multi_cell(0, 10, body)
        self.ln()

# Create instance of FPDF class
pdf = PDF()

# Add a page
pdf.add_page()

# Set title
pdf.chapter_title('Business Insights')

# Set body
insights = """
The business insights provide a comprehensive view of performance across various dimensions. South America emerges as the leading region in terms of total sales, generating a revenue of $219,352.56. In terms of product preferences, the most purchased products include Product IDs P059, P054, and P029, with quantities reaching up to 46 units each. Additionally, the most popular categories are Books, Home Dcor, and Electronics, which collectively dominate customer preferences.

The sales trends show significant growth in revenue over the months of 2024, with monthly totals peaking in July ($71,366.39) and consistent performance observed throughout the year. Comparatively, the total revenue for 2023 was minimal ($3,769.52), indicating the business has significantly scaled up operations in 2024, achieving an annual total of $686,226.04.

High-value customers play a pivotal role in driving revenue. The top customer, C0141, contributed $10,673.87, with others, including C0054 and C0065, closely following. These insights highlight loyal customers' impact and opportunities for targeted retention strategies.

This analysis demonstrates strong business growth, customer engagement, and product-market alignment, providing a foundation for further optimization and scaling strategies. Furthermore, peak months for customer signups is March.
"""

pdf.chapter_body(insights)

# Save the PDF to a file
pdf.output('business_insights.pdf')

print("PDF created successfully.")
```
